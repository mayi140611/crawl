```python
from pyquery import PyQuery as pq
```

```python
with open('39疾病百科_完善的疾病数据库.html', 'r', encoding='utf8') as f:
    html = f.read()
d = pq(html)
```

```python
mainItems = d('#map > div > dl.mapItem')
#部位列表[[(一级分类,url),(二级分类,url),...],[(一级分类,url),(二级分类,url),...],...]
buwei_list = []
for i in range(len(mainItems)):
    l1 = []
    a = mainItems.eq(i).find('dt > a.bold').eq(0) 
    l1.append((a.text(),a.attr.href))
#     print(a.text(),a.attr.href)
    lis = mainItems.eq(i).find('dd > div:nth-child(1) > ul > li > span.left > a')
    for ii in range(len(lis)):
#         print(lis.eq(ii).text(),lis.eq(ii).attr.href)
        l1.append((lis.eq(ii).text(),lis.eq(ii).attr.href))
    buwei_list.append(l1)
```

## 获取每个二级分类下的疾病和症状列表

```python
import requests
from pyquery import PyQuery as pq
```

```python
# count = 0
def get_index(url, jibing_list):
    '''
    获取url页面上所有的疾病和疾病链接
    '''
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.79 Safari/537.36 Maxthon/5.2.1.6000'}
    html = requests.get(url,headers=headers)
    d = pq(html.text)
    las = d('.res_list > dl > dt > h3 > a')
    for i in range(len(las)):
#         print(las.eq(i).text(),las.eq(i).attr.href)
        jibing_list.append((las.eq(i).text(),las.eq(i).attr.href))
    last_page = d('a.sp-a:contains("下页")')
    if last_page.eq(0):
#         global count
#         if count>2:
#             return
#         count += 1
        url = 'http://jbk.39.net{}'.format(last_page.eq(0).attr.href)
        print(url)
        get_index(url, jibing_list)    
```

```python
#分类及其下疾病列表[[(二级分类,url),(疾病,url),...],[(二级分类,url),(疾病,url),...],...]
all_jibing_list = []
all_zz_list = []
for t in buwei_list:
    for i in range(len(t)):
        if i>0:
            try:
                print(t[i])
                jibing_list = []
                zz_list = []
                jibing_list.append(t[i])
                url = '{}_t1'.format(t[i][1][:-1])
                get_index(url, jibing_list)
                zz_list.append(t[i])
                url = '{}_t2'.format(t[i][1][:-1])
                get_index(url, zz_list)
                all_jibing_list.append(jibing_list)
                all_zz_list.append(zz_list)
            except:
                print(url)
```

```python
import pickle
with open('all_jibing_list.pkl', 'wb') as f:
    pickle.dump(all_jibing_list,f)
with open('all_zz_list.pkl', 'wb') as f:
    pickle.dump(all_zz_list,f)

with open('buwei_list.pkl', 'wb') as f:
    pickle.dump(buwei_list,f)
```
