```python
mainItems = d('#map > div > dl.mapItem')
#部位列表[[(一级分类,url),(二级分类,url),...],[(一级分类,url),(二级分类,url),...],...]
buwei_list = []
for i in range(len(mainItems)):
    l1 = []
    a = mainItems.eq(i).find('dt > a.bold').eq(0) 
    l1.append((a.text(),a.attr.href))
#     print(a.text(),a.attr.href)
    lis = mainItems.eq(i).find('dd > div:nth-child(1) > ul > li > span.left > a')
    for ii in range(len(lis)):
#         print(lis.eq(ii).text(),lis.eq(ii).attr.href)
        l1.append((lis.eq(ii).text(),lis.eq(ii).attr.href))
    buwei_list.append(l1)
```

### 2、获取疾病的典型症状列表

```python
from utils.pymongo_wrapper import pymongo_wrapper as pmw
```

```python
client = pmw()
```

```python
table = client.get_collection('jiankang39','diseases')
```

```python
import requests
from pyquery import PyQuery as pq
def parse_disease(url):
    '''
    解析疾病页面
    '''
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.79 Safari/537.36 Maxthon/5.2.1.6000'}
    html = requests.get(url,headers=headers)
    html.encoding = 'gbk'
    d = pq(html.text)
    dds = d('div.info > ul > li')
    for i in range(len(dds)):
        label = dds.eq(i)('i').eq(0).text()
        s = dds.eq(i).text()    
        if s.endswith('[详细]'):
            s = s[:-4].strip()
        ss = [i.strip() for i in s.split('：')]
        content = dds.eq(i)('a')
        if content and ss[0] == '典型症状':
            ll = list()
            for ii in range(len(content)):
                if content.eq(ii).attr.title:
                    ll.append((content.eq(ii).attr.title,content.eq(ii).attr.href))
            return ll
    return None
```

```python
parse_disease('http://jbk.39.net/ptt/')
```

```python
all_zz_list2 = set()
for t in client.find_all(table,fieldlist=['url']):
    try:
        all_zz_list2 = all_zz_list2.union(parse_disease(t['url']))
    except Exception as e:
        print(e)
        print(t)
```
