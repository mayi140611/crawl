```python
#设置代理
import requests

PROXY_POOL_URL = 'http://localhost:5555/random'

def get_proxy():
    try:
        response = requests.get(PROXY_POOL_URL)
        if response.status_code == 200:
            print('using proxy',response.text)
            return {'http':'http://'+response.text}
    except ConnectionError:
        return None
```

```python
#爬取文章正文部分
count = 0
proxy = None
for item in list1:
    try:
        rqw = requests_wrapper.requests_wrapper(item[1])
        print(count, rqw.get_response(proxies=proxy).status_code)
        text = rqw.get_text()
    except Exception as e:
        print(e)
        try:
            rqw = requests_wrapper.requests_wrapper(item[1])
            proxy = get_proxy()
            if proxy:
                print(count, rqw.get_response(proxies=proxy).status_code)
                text = rqw.get_text()
            else:
                print('crawl failure：获取代理失败',item)
                continue
        except Exception as e:
            print(e)
            print('crawl failure',item)
            continue
        
    pqw = pyquery_wrapper.pyquery_wrapper(text)
    count += 1
    root = pqw.Root    
    dict1 = dict()
    dict1['tags']=list()
    dict1['name']=item[0]
    dict1['url']=item[1]
    dict1['time']=item[2]
    tags = root('.blog_tag>h3')
    if tags:
        for i in range(len(tags)):
            dict1['tags'].append(tags.eq(i).text())
    dict1['blogclass']=list()
    blogclass = root('.blog_class>a')
    if blogclass:
        for i in range(len(blogclass)):
            dict1['blogclass'].append(blogclass.eq(i).text())
    content = root('.articalContent').eq(0).html()
    dict1['content'] = content
    chan.insert_one(dict1)
```
